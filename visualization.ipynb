```python
# An overview of employment of different states in Malaysia from 1982 to 2020

fig = go.Figure()

for state in df_emp["State/Country"]:
    fig.add_trace(
        go.Scatter(
            x = df_emp[df_emp["State/Country"] == state]["Year"],  
            y = df_emp[df_emp["State/Country"] == state]["Employed"],
            mode = 'lines+markers',
            name = state
        )
    )


fig.update_layout(
    title="Employment Over Time",
    xaxis_title="Year",
    yaxis_title="Population",
    template="plotly_white",
    hovermode="x unified"
)
```
